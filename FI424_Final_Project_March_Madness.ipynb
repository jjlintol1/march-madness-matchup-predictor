```python
# Create a list of the file names (named in "YYYY_YY.csv" format in Google Drive) - Courtesy of ChatGPT
# Initialize an empty list to store the years
years_list = []

# Start year
start_year = 1992

# End year
end_year = 2018

# Loop through the years and add them to the list in the desired format
for year in range(start_year, end_year + 1):
    formatted_year = f"{year}_{str(year + 1)[-2:]}"  # Get the last two digits of the next year
    years_list.append(formatted_year)

# Print the list of years
print(years_list)
```

### Import Pandas, read in data and begin manipulation

```python
import pandas as pd
```

```python
# Create a list of DataFrame instances and concatenate them
dfs = []

for year in years_list:
  current_df = pd.read_csv(f"/content/drive/MyDrive/MarchMadness/{year}.csv")
  # Label each dataset by year
  current_df["Year"] = int(year.split("_")[0]) + 1
  # The sports reference data had "NCAA" tags on tournament qualifying schools. I remove this here
  current_df["School"] = current_df["School"].str.replace("NCAA", "").str.strip()
  print(year)
  print(current_df.columns)
  dfs.append(current_df)

# Combine all the data to one DataFrame
df = pd.concat(dfs)
```

#### Combine the team stats data with the matchup data to make it neural network ready

```python
# Reset index of original DataFrame to make mapping easier
df.reset_index(drop=True, inplace=True)
```

```python
# Use the mapping from the top of the notebook to create a key to match the data on
for index, row in df.iterrows():
  print(index, row["School"])
  # This code creates a key to match the team stats df to the matchups df
  if row["School"] in mapping.keys():
    df.loc[index, "team_match"] = mapping[df.loc[index, "School"]]
  else:
    df.loc[index, "team_match"] = "N/A"
```

```python
# List of teams there is no mapping for (it repeats the teams because it lists for every season)
for index, row in df.iterrows():
  if row["team_match"] == "N/A":
    print(row["School"])
```

```python
df.columns, matchups_df.columns
```

```python
# Join the DataFrame instances based on the newly created keys and add team stats to the matchup data
for index, row in matchups_df.iterrows():
  print(f"Index {index} of {len(matchups_df)}")
  # Iterate through all columns and add them to the matchup data
  for col in df.columns:
    # Find the current column for "team" in matchup data and add it to matchup data
    team_1 = df[(df["Year"] == row["year"]) & (df["team_match"] == row["team"])].loc[:, col]
    if len(team_1) > 0:
      matchups_df.loc[index, f"team_1_{col.lower()}"] = team_1.iloc[0]
    # Find the current column for "team_2" in matchup data and add it
    team_2 = df[(df["Year"] == row["year"]) & (df["team_match"] == row["team_2"])].loc[:, col]
    if len(team_2) > 0:
      matchups_df.loc[index, f"team_2_{col.lower()}"] = team_2.iloc[0]
```

#### Now that we have removed null values, let's take the final steps to get the data neural network ready

```python
# Create a column indicating the winner of each matchup
for index, row in matchups_df.iterrows():
  score = matchups_df.loc[index, "score"]
  score_2 = matchups_df.loc[index, "score_2"]
  # If team 2 wins, winner will be 1 else 0 for team 1
  matchups_df.loc[index, "winner"] = 1 if score_2 > score else 0
```

```python
# Create a list of object columns to drop (none of these need to be encoded because they do not contrubute to outcome)
drop_columns = [col for col in matchups_df.columns if matchups_df[col].dtype not in ['int64', 'float64']]
drop_columns
```

```python
matchups_df.head()
```

```python
# Manually adding more drop columns that are numeric but are either irrelevant or give away the outcome
additional_drop_columns = [
    "year",
    "region_number",
    "seed",
    "score",
    "score_2",
    "seed_2",
    "team_1_rk",
    "team_2_rk",
    "team_1_year",
    "team_2_year",
    "winner"
]
```

```python
# Combine drop columns
drop_columns.extend(additional_drop_columns)
drop_columns
```

```python
# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Create X and Y with columns dropped
X = matchups_df.drop(axis=1, columns=drop_columns)
y = matchups_df["winner"]

matchups_df.shape, len(drop_columns), X.shape, y.shape
```

```python
# Scale all values to be between 0 and 1
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Split data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Check train and test data
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## 1. Create Neural Network

The data is ready. We can now create the neural network.

```python
# Imports
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
tf.random.set_seed(42)

# Create model
model = Sequential()
model.add(Dense(32, input_dim=49, activation="relu"))
model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

# Early stopping callback
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=10
)

# Compile model
model.compile(
    optimizer=Adam(),
    loss=BinaryCrossentropy(),
    metrics=["accuracy"]
)
```

```python
# Train model
history = model.fit(
    X_train,
    y_train,
    epochs=100,
    callbacks=[early_stopping],
    validation_split=0.2
)
```

```python
# Evaluate model on test data
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss on test data: ", loss)
print("Accuracy on test data: ", accuracy)
```
